```python
with open("gym_member.txt", "r")as file:
  content = file.read()
  print(content)
```

```python
with open("gym_member.txt", "r")as file:
  for line in file:
    print(line.strip())
```

```python
with open("gym_member.txt", "a")as file:
    file.write("Rohit|Crossfit|300\n")

print("Member added!")

with open("gym_member.txt", "r")as file:
   for line in file:
    print(line.strip())

```

```python
import json

gym_data = {
    "gym_name" : "Fit_india",
    "members": {
        "Rahul" :{"Plan" : "HIIT", "Fee" : 2000},
        "Priya" :{"Plan" :"Cardio", "Fee" : 1500},
        "Raj" : {"Plan" : "Crossfit", "Fee" : 2300}
    }

}

with open("gym_data.json", "w")as file:
  json.dump(gym_data,file, indent=4)

print("Json file created!")
```

```python
with open("gym_data.json", "r")as file:
  for line in file:
    print(line.strip())
```

```python
with open("gym_data.json", "r")as file:
  data = json.load(file)

# Use it like a normal dictionary!
print(data["gym_name"])
print(data["members"]["Rahul"]["Plan"])

# Loop through all members:
for name,details in data["members"].items():
  print(f"{name} |{details["Plan"]} |{details["Fee"]}")
```
