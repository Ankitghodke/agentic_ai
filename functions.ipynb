```python
def add(a,b):
  return(a+b)

def subtract(a,b):
  return(a-b)

def multiply(a,b):
  return(a*b)

print(add(10,5))
print(subtract(10,5))
print(multiply(10,5))
```

```python
def greet(name, role="Member"):
  print(f"Hello {name}! You are a {role}")

greet("Rahul")
greet("Priya", "Trainer")
```

```python
def gym_members(*members):
    for member in members:
      print(f"{member} is a member!")

gym_members("Rahul")
gym_members("Rahul", "Priya")
gym_members("Rahul", "Priya", "Amit")
```

```python
def member_details(**details):
  for key, value in details.items():
      print(f"{key}: {value}")

member_details(name = "Rahul", Plan = "HIIT", Fees = 2000)
member_details(name = "Raj", Plan = "Cardio", Fees = 2200, age = 22)
```

```python
def add_member(members, name, plan, fee):
  members[name] = {"plan" : plan, "fee" : fee}
  print(f"{name} added!")

def show_members(members):
  for name, details in members.items():
    print(f"{name} | {details["plan"]} | {details["fee"]}")

def find_members(members,name):
  if name in members:
    print(f"found:{name} | {members[name] ["plan"]}")

  else:
    print(f"{name} not found!")

gym = {}
add_member(gym, "Rahul", "HIIT", 2000)
add_member(gym,"Raj", "Cardio", 1500)
show_members(gym)
find_members(gym, "Rahul")
find_members(gym, "Priya")
```
