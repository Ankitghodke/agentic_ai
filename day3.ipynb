```python
gym_members = {
    "Rahul" : {"age":23, "plan":"HIIT", "fee":2000},
    "Priya" : {"age":22, "plan":"Cardio", "fee":1500},
    "Amit" : {"age":24, "plan":"Strength", "fee":2000}
}

# Access Rahul's age
print(gym_members["Rahul"] ["age"])

# Access Priya's plan
print(gym_members ["Priya"] ["plan"])

# loop through all members
for name,details in gym_members.items():
  print(f"{name} |{details["age"]} | {details["plan"]} | {details["fee"]}")
```

```python
class Gym:
  def __init__(self):
    self.members = {}

  def add_members(self,name,plan,fee):
    self.members[name] = {"plan" : plan, "fee": fee}
    print(f"{name} added!")

  def show_members(self):
    for name,details in self.members.items():
      print(f"{name} | {details["plan"]} | {details["fee"]}")

  def find_members(self, name):
    if name in self.members:
      print(f"found:{name} | {self.members[name] ["plan"]}")
    else:
      print(f"{name} not found!")

g = Gym()
g.add_members("Rahul", "HIIT", 2000)
g.add_members("Priya", "Cardio", 1500)
g.add_members("Amit", "Strength", 2500)
g.show_members()
g.find_members("Priya")
g.find_members("Vikram")
```
