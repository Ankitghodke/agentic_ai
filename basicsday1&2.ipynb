```python
class person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
  def showdetails(self):
    print(f"name:{self.name},age:{self.age}")

class engineer(person):
  def __init__(self,name,age,skill):
    super().__init__(name, age)
    self.skill=skill

class AIengineer(engineer):
  def __init__(self,name,age,skill,tools):
    super(). __init__(name,age,skill)
    self.tools=tools

  def show_tools(self):
    print(f"{self.name} uses {self.tools}")

  def show_skill(self):
    print(f"{self.name} knows {self.skill}")

eng1=AIengineer("Rahul",23,"python","Cloude API")
eng1.show_tools()
eng1.showdetails()
eng1.show_skill()
```

```python
class vehicle:
  def __init__(self,brand,speed):
    self.brand = brand
    self.speed = speed
  def show_all(self):
    print(f"{self.brand} | {self.speed}")

class car(vehicle):
  def __init__(self,brand,speed,fuel_type):
    super().__init__(brand,speed)
    self.fuel_type = fuel_type

class ElectricCar(car):
  def __init__(self,brand,speed,fuel_type,battery_capicity):
    super().__init__(brand,speed,fuel_type)
    self.battery_capicity = battery_capicity

  def show_all(self):
    print(f"{self.brand} | {self.speed} | {self.fuel_type} | {self.battery_capicity}")

ec = ElectricCar ("Tesla",250,"Electric","100kwh")
ec.show_all()
```

```python
class person:
  def __init__(self,name,age):
    self.name = name
    self.age = age
class Employee(person):
  def __init__(self,name,age,company,salary):
    super().__init__(name,age)
    self.company=company
    self.salary=salary
class Manager(Employee):
  def __init__(self,name,age,company,salary,team_size):
    super().__init__(name,age,company,salary)
    self.team_size=team_size

  def showall(self):
    print(f"{self.name} | {self.age} | {self.company} | {self.salary} |{self.team_size}")

m = Manager("Rahul",24,"Google",15000,10)
m.showall()
```

```python
class member:
  def __init__(self,name,age,plane):
    self.name = name
    self.age = age
    self.plane = plane

class gym:
  def __init__(self):
   self.member = []

  def add_member(self,member):
    self.member.append(member)
    print(f"{member.name} added to gym!")

  def show_member(self):
    for member in self.member:
      print(f"{member.name} | {member.age} | {member.plane}")

  def find_member(self,name):
    for member in self.member:
      if member.name == name:
         print(f"found:{member.name}")
         break
    else:
       print("Member not found")


work = gym()
work.add_member(member("ankit",16,"HIIT"))
work.add_member(member("Raj",17,"Cardio"))
work.add_member(member("Shrikant",16,"Strength"))
work.show_member()
work.find_member ("Vikram")
work.find_member ("Raj")

```
