```python

def calcluate_bmi(weight, height):
  return weight / (height **2)

def bmi_category(bmi):
  if bmi < 18.5:
    return "Underweight"
  elif bmi <= 24.9:
    return "Normal"
  else:
    return "Overweight"

people = [
    {"name" : "Rahul", "weight" : 50, "height" : 1.60},
    {"name" : "Priya", "weight" : 45, "height" : 1.55},
    {"name" : "Rohit", "weight" : 49, "height" : 1.67}
]

for person  in people:
  bmi = calcluate_bmi(person["weight"], person["height"])
  category = bmi_category(bmi)
  print(f"{person["name"]} | BMI: {bmi:.1f} | {category}")

```

```python
def weekly_dose(daily_dose,days=7):
  return daily_dose*days

def cycle_cost(dose,price_per_mg):
  return dose*price_per_mg

dose = weekly_dose(20,10)
print(f"Weeklydose : {dose}mg")
print(f"cost: ₹{cycle_cost(dose, 800)}")

dose = weekly_dose(20)
print(f"Weeklydose : {dose}mg")
print(f"cost: ₹{cycle_cost(dose, 800)}")



```

```python
import json

def load_contacts():
  try:
    with open("contacts.json", "r")as file:
      return json.load(file)
  except FileNotFoundError:
    return {}

def save_contacts(contacts):
  with open("contacts.json", "w")as file:
    json.dump(contacts,file, indent = 4)

def add_contacts(contacts, name, number):
  contacts[name] = number
  save_contacts(contacts)
  print(f"{name} added!")

def show_contacts(contacts):
  for name,details in contacts.items():
    print(f"{name} | {details}")

def find_contacts(contacts,name):
  if name in contacts:
    print(f"found : {name} | {contacts[name]}")
  else:
    print(f" {name}Name not found!")

def delete_contacts(contacts, name):
  if name in contacts:
    del contacts[name]
    save_contacts(contacts)
    print(f"{name} deleted!")
  else:
    print(f"{name} not found!")

contacts = load_contacts()
add_contacts(contacts, "Mom", "9876543210")
add_contacts(contacts, "Dad", "9876543211")
show_contacts(contacts)
find_contacts(contacts, "Mom")
delete_contacts(contacts, "Dad")
show_contacts(contacts)
```

```python

import json

def load_members():
  try:
    with open("members.json", "r")as file:
      return json.load(file)
  except FileNotFoundError:
    return {}

def save_members(members):
  with open("members.json", "w")as file:
    json.dump(members,file , indent = 4)

def add_members(members,name,plan,fee):
  members[name] = {"plan" : plan, "fee" : fee}
  save_members(members)
  print(f"{name} added!")

def show_members(members):
  for name,details in members.items():
    print(f"{name} | {details}")

def find_members(members,name):
  if name in members:
    print(f"{name} member found")
  else:
    print(f"{name} not found")


members = load_members()

while True:
  print("\n1. Add member")
  print("2. show all members")
  print("3. Find member")
  print("4. Quit")

  choice = input("Enter the choice: ")

  if choice == "1":
      name = input("Enter the name: ")
      plan = input("Enter the plan: ")
      fee = input("Enter the fee: ")
      add_members  (members, name, plan, fee)

  elif choice == "2":
      show_members(members)

  elif choice == "3":
      name = input("Enter the name to find: ")
      find_members(members, name)

  elif choice =="4":
     print("Goodbye!")
     break

  else:
     print("Invalid choice!")
```
