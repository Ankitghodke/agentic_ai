```python
countries = {
   "India": "New Delhi",
    "USA": "Washington DC",
    "France": "Paris",
    "Australia": "Canberra",
    "Japan": "Tokyo"
}

for country,capital in countries.items():
  print(f"{country} → {capital}")

print("India" in countries)
print("Mars" in countries)
```

```python
ufc_fighters = {
    "Jon_Jones" : {"weight" : "Heavyweight", "fights" : 28, "win_rate" : "95%"},
    "Islam_Makhachev" : {"weight" : "Lightweight", "fights" : 26, "win_rate" : "92%"},
    "Alex_Pereria" : {"weight" : "LightHeavyweight", "fights" : 12, "win_rate" : "88%"}
}

for name,details in ufc_fighters.items():
  print(f"{name} |{details["weight"]} | {details["fights"]} | {details["win_rate"]}")

print(f"Jon_Jones win rate:{ufc_fighters['Jon_Jones']['win_rate']}")
```

```python
cycle = {
    "denabol" :{"Quantity":"20mg","Price":800},
    "Trenbalon" :{"Quantity":"15mg","Price":2000},
    "Test" :{"Quantity":"25mg","Price":2200},
    "Anavar" :{"Quantity":"10mg","Price":2000},
    "deca" :{"Quantity":"25mg","Price":2400}
}

try:
  name = input("Enter the Roid:")
  print(f"info :{cycle[name]}")
except KeyError:
  print(f"{name} not found product!")
finally:
  print("Search completed")
```

```python
try:
  number = float(input("Enter the number:"))
  operator = input("Enter the operator (+, -, *, / ):")
  number1 = float(input("Enter the 2nd  number:"))

  if operator == "+":
    result = number + number1
  elif operator == "-":
    result = number - number1
  elif operator == "*":
    result = number * number1
  elif operator == "/":
    result = number / number1

  print(f"result : {result}")


except ValueError:
  print("Please enter number only!")

except ZeroDivisionError:
  print("You can't divide by zero!")

finally :
  print("Calcluation Over!")
```
