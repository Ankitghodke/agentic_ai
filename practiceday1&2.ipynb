```python
class student:
  def __init__(self,name,age,grade):
    self.name = name
    self.age = age
    self.grade = grade

class school:
  def __init__(self):
    self.student = []

  def add_student(self,student):
    self.student.append(student)
    print(f"{student.name} added!")

  def show_student(self):
   for std in self.student:
    print(f"{std.name} | {std.age} | {std.grade}")

  def find_student(self,name):
      for std in self.student:
          if std.name == name:
             print(f"found {std.name}")
             break
      else:
          print("student not found!")

std1 = school()
std1.add_student(student("Rahul",16,"A"))
std1.add_student(student("Ramu",17,"C"))
std1.show_student()
std1.find_student("Ankit")
std1.find_student("Rahul")




```

```python
class Animal:
  def __init__(self,name,age):
    self.name = name
    self.age = age
    print(f"{}")

class Dog(Animal):
  def __init__(self,name,age,breed):
    super().__init__(name,age)
    self.breed = breed

class GuideDog(Dog):
  def __init__(self,name,age,breed,trained_for):
    super().__init__(name,age,breed)
    self.trained_for = trained_for

  def show_all(self):
    print(f"{self.name} | {self.age} | {self.breed} | {self.trained_for}")

kutta = GuideDog("Arush",3,"Pagal_Kutta","Blind_assistance")
kutta.show_all()
```

```python
class Movies:
  def __init__(self,title,director,year):
    self.title = title
    self.director = director
    self.year = year

class Cinemaa:
  def __init__(self):
    self.Movies = []

  def add_Moviess(self, Movies):
    self.Movies.append(Movies)
    print(f"{Movies.title} | {Movies.director} | {Movies.year}")

pictures = Cinemaa()

pictures.add_Moviess(Movies("KGF","Pappu",2018))
pictures.add_Moviess(Movies("RRR","Tappu",2019))
```

```python
class Employee:
  def __init__(self,name,department,salary):
    self.name = name
    self.department = department
    self.salary = salary

class Company:
  def __init__(self):
    self.Employee = []

  def add_employee(self,Employee):
    self.Employee.append(Employee)
    print(f"{Employee.name} added!")

  def show_employee(self):
    for Emp in self.Employee:
      print(f"{Emp.name} | {Emp.department}")

c = Company()
c.add_employee(Employee("Ankit","IT",20000))
c.add_employee(Employee("Rahul","HR",15000))
c.show_employee()

```

```python
class book:
  def __init__(self,title,author,price):
    self.title = title
    self.author = author
    self.price = price

class Bookstore:
  def __init__(self):
    self.book = []

  def add_book(self,book):
    self.book.append(book)
    print(f"{book.title} added!")

  def show_books(self):
    for ketab in self.book:
        print(f"{ketab.title} | {ketab.author} | {ketab.price} Showed!")

  def find_book(self,title):
    for ketab in self.book:
        if ketab.title == title:
           print(f"{ketab.title} ketab melgaya bhavva")
           break
    else:
           print(f"No book found")

b = Bookstore()
b.add_book(book("Harry Potter", "J.K Rowling", 500))
b.add_book(book("Python Basics", "Guido", 300))
b.show_books()
b.find_book("Harry Potter")

```
